##  3.8.2   Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.

#SQL QUERY  

SELECT `STATE/UT`, YEAR, DISTRICT, MURDER  
FROM (  
    SELECT   
        `STATE/UT`,   
        YEAR,    
        DISTRICT,   
        MURDER,  
        ROW_NUMBER() OVER(PARTITION BY `STATE/UT`, YEAR ORDER BY MURDER DESC) AS rn   
    FROM `01_district_wise_crimes_committed_ipc_2001_2012   
) AS ranked   
WHERE rn = 1;

## 3.8.4 Use appropriate graphs to show your data (the result of 3.8.3)

```python
sortdf=newdf.groupby('STATE/UT')['MURDER'].sum().reset_index().sort_values(by='MURDER',ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='STATE/UT', y='MURDER', data=sortdf.head(10))
plt.title('Top 10 States with Highest Total Murders')
plt.xlabel('State/UT')
plt.ylabel('Total Murders')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
top_states = sortdf['STATE/UT'].head(5).tolist()
plt.figure(figsize=(12, 6))
for state in top_states:
    state_data = newdf[newdf['STATE/UT'] == state]
    sns.lineplot(x='YEAR', y='MURDER', data=state_data, label=state)
plt.title('Year-wise Comparison of Murders in Top States')
plt.xlabel('Year')
plt.ylabel('Total Murders')
plt.legend(title='State/UT', loc='upper right')
plt.show()
```

```python
# Heatmap of Murders by State and Year
murders_heatmap = newdf.pivot_table(index='STATE/UT', columns='YEAR', values='MURDER', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(murders_heatmap, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Heatmap of Murders by State and Year')
plt.xlabel('Year')
plt.ylabel('State/UT')
plt.show()
```
